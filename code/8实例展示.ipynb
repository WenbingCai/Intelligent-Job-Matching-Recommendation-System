```python
data = pd.read_csv(r"./eg_data.csv")
data = data.drop(['user_id'], axis = 1)
data = data.drop(['jd_no'], axis = 1)
data = data.drop(['Unnamed: 0'], axis = 1)
data = data.dropna(axis = 0,subset =['Unstructured_similarity'])
data = data.fillna(0)
X = data
```

```python
data.isnull().sum()
```

```python
y_shat = clf_s_r.predict_proba(X)
y_dhat = clf_d_xg.predict_proba(X)
S_hat = y_shat[:,1]
D_hat = y_dhat[:,1]
```

```python
score = 0.82 * S_hat + (1-0.82)* D_hat
pro_s = S_hat
```

```python
data_final = pd.read_csv('eg_data.csv')
data_final = data_final.dropna(axis = 0,subset = ['Unstructured_similarity'])
data_final.loc[:,'score']=score
data_final.loc[:,'pro_s']=pro_s
data_final.to_csv('eg_data_final.csv',encoding='utf_8_sig')
data_final = data_final.sort_values(by='score',axis=0,ascending=False,inplace=False,na_position='last')
data_final
```
