```python
def normal_leven(str1, str2):
  len_str1 = len(str1) + 1
  len_str2 = len(str2) + 1
  # 创建矩阵
  matrix = [0 for n in range(len_str1 * len_str2)]
  #矩阵的第一行
  for i in range(len_str1):
    matrix[i] = i
  # 矩阵的第一列
  for j in range(0, len(matrix), len_str1):
    if j % len_str1 == 0:
      matrix[j] = j // len_str1
  # 根据状态转移方程逐步得到编辑距离
  for i in range(1, len_str1):
    for j in range(1, len_str2):
      if str1[i-1] == str2[j-1]:
        cost = 0
      else:
        cost = 1
      matrix[j*len_str1+i] = min(matrix[(j-1)*len_str1+i]+1,
                    matrix[j*len_str1+(i-1)]+1,
                    matrix[(j-1)*len_str1+(i-1)] + cost)
 
  return matrix[-1]  
```

```python
datas['jd_sub_type'] = datas['jd_sub_type'].astype(str)
datas['desire_jd_industry_id'] = datas['desire_jd_industry_id'].astype(str)
for i in range(len(datas['jd_sub_type'])):
    datas['jd_sub_type'][i] = datas['jd_sub_type'][i].replace('/','')
for i in range(len(datas['desire_jd_industry_id'])):
    datas['desire_jd_industry_id'][i] = datas['desire_jd_industry_id'][i].replace('/','')
text1=datas['jd_sub_type']
text2=datas['desire_jd_industry_id']
text1=text1.values
text2=text2.values
```

```python
data=pd.read_csv(r"C:\Users\caiwenbing\Desktop\data(1).csv")
```

```python
for i in range(len(text1)):
    simi2 = normal_leven(text1[i], text2[i])
    data.loc[i,'String_similarity']=simi2
```
