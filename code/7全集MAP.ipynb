```python
import pandas as pd
import numpy as np
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
data_final = pd.read_csv('data_final.csv')
data_final = data_final.sort_values(by='score',axis=0,ascending=False,inplace=False,na_position='last')
data_final
```

```python
def MAP():
    
    a = data_final['user_id'].values
    user = []
    for i in a:
        if i in user:
            continue
        else:
            user.append(i)
            
    hit_s = []
    hit_d = []
    AP_d_list = []
    AP_s_list = []

    for i in user:
        recommend = data_final[data_final['user_id']==i]
        recommend = recommend.sort_values(by='score',axis=0,ascending=False,inplace=False,na_position='last')
        hit_s = recommend['satisfied'].to_list()
        hit_d = recommend['delivered'].to_list()
        
        AP_s = 0
        AP_d = 0
        
        for j in range(len(recommend)): 
            AP_s = AP_s + hit_s[j]*sum(hit_s[:j+1])/(j+1)        
            AP_d = AP_d + hit_d[j]*sum(hit_d[:j+1])/(j+1) 
        n_s = sum(hit_s[:])
        n_d = sum(hit_d[:])
        if n_s != 0 and n_d != 0:
            AP_s_list.append(AP_s/n_s)
            AP_d_list.append(AP_d/n_d)

    MAP_d = sum(AP_d_list)/len(user)
    MAP_s = sum(AP_s_list)/len(user)
    MAP = 0.7*MAP_s + 0.3*MAP_d

    return MAP
```

```python
MAP()
```
